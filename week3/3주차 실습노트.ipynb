```python
#7 recursion

def fib(n):
  if n == 1 or n == 2:
    return 1
  else:
    return fib(n - 1) + fib(n - 2)
  
print(fib(5))
print(fib(10))
print(fib(35))
```

```python
#8 memoization

def fib(n, m):
  if n == 1 or n == 2:
    return 1
  else:
    if m[n-1] == 0:
      m[n-1] = fib(n-1, m)
    if m[n-2] == 0:
      m[n-2] = fib(n-2, m)
    return m[n-1] + m[n-2]

fibn = 5
m = [ 0 for i in range(0, fibn)]
print(fib(fibn, m))

fibn = 10
m = [ 0 for i in range(0, fibn)]
print(fib(fibn, m))

fibn = 35
m = [ 0 for i in range(0, fibn)]
print(fib(fibn, m))

fibn = 5
m = [ 0 for i in range(0, fibn)]

def fib(n, m):
  if n == 1 or n == 2:
    return 1
  else:
    if m[n - 1] == 0:
      m[n - 1] = fib(n - 1, m)
    if m[n - 2] == 0:
      m[n - 2] = fib(n - 2, m)
    return m[n - 1] + m[n - 2]
```

```python
def fact(n):
  result = 1;

  for a in range(2, n + 1):
    result *= a

  return result

n = int(input('input the number : '))
print(fact(n))   
```

```python
def intsum(*ints):
  sum = 9
  for num in ints:
    sum += num
  return sum

print(intsum(1, 2, 3))
print(intsum(5, 7, 9, 10))
print(intsum(1, 3, 5, 7, 9, 11, 13, 15))
```

```python
def calcstep(a, b, step = 1):
  sum = 0
  for num in range(a, b + 1, step):
    sum += num
  return sum

print(calcstep(1, 10, 2))


calcstep(1, 10, 2)
calcstep(a = 1, b = 10, step = 2)
```
