```python
poet = """Do not go gentle into that good night,
  Old age should burn and rave at close of day;
  Rage, rage against the dying of the light."""

alpha = dict()
for c in poet:
  if c.isalpha() == False:
    continue
  c = c.lower()
  if c not in alpha:
    alpha[c] = 1
  else:
    alpha[c] += 1
  
print(alpha)

alpha = dict()
```

```python
#zip(컬렉션1, 컬렉션2, ...)

dayofweek = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"] 
food = ["Pasta", "Steak", "Cheese", "Toast"]
menu = zip(dayofweek, food) # 두 리스트를 병합하여 새 리스트 생성 
for d, f in menu:
    print("The menu of %s : %s" % (d, f))

menu = dict(zip(dayofweek, food))
print(menu)
menu = list(zip(dayofweek, food))
print(menu)
```

```python
#filter(필터링 조건 함수, 대상 리스트)

def flunk(s):
  return s < 60

score = [45, 89, 72, 53, 94]
for s in filter(flunk, score):
  print(s)
```

```python
#map(변환 함수, 대상 리스트)

def half(s):
  return s / 2

score = [45, 89, 72, 53, 94]
for s in map(half, score):
  print(s, end = ", ")
```

```python
def total(s, b):
  return s + b

score = [45, 89, 72, 53, 94]
bonus = [2, 3, 0, 0, 5]
for s in map(total, score, bonus):
  print(s, end = ', ')
```
